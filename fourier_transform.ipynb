```python
import librosa
import IPython.display as ipd

D3_PATH = "audio/D3.mp3"
F3_PATH = "audio/F3.mp3"
C4_PATH = "audio/C4.mp3"
E4_PATH = "audio/E4.mp3"

d3_y, sr = librosa.load(D3_PATH)
f3_y, sr = librosa.load(F3_PATH)
c4_y, sr = librosa.load(C4_PATH)
e4_y, sr = librosa.load(E4_PATH)

min_len = min(len(d3_y), len(f3_y), len(c4_y), len(e4_y))

d3_y = d3_y[:min_len]
f3_y = f3_y[:min_len]
c4_y = c4_y[:min_len]
e4_y = e4_y[:min_len]

chord_y = d3_y + f3_y + c4_y + e4_y

ipd.display(ipd.Audio(d3_y, rate=sr))
ipd.display(ipd.Audio(f3_y, rate=sr))
ipd.display(ipd.Audio(c4_y, rate=sr))
ipd.display(ipd.Audio(e4_y, rate=sr))

```

```python
from plot import plot_audio_signal

min_len = min(len(d3_y), len(f3_y), len(c4_y), len(e4_y))

d3_y = d3_y[:min_len]
f3_y = f3_y[:min_len]
c4_y = c4_y[:min_len]
e4_y = e4_y[:min_len]

chord_y = d3_y + f3_y + c4_y + e4_y

plot_audio_signal(d3_y[sr:sr+1000], sr=sr, title="D3 note")
plot_audio_signal(f3_y[sr:sr+1000], sr=sr, title="F3 note")
plot_audio_signal(c4_y[sr:sr+1000], sr=sr, title="C4 note")
plot_audio_signal(e4_y[sr:sr+1000], sr=sr, title="E4 note")
plot_audio_signal(chord_y[sr:sr+1000], sr=sr, title="Chord")

ipd.display(ipd.Audio(chord_y, rate=sr))
```
